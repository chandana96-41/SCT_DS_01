```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#BAR PLOT
# "How does the Happiness Score vary across different regions?"
#if there are noticeable differences in the average happiness levels of countries belonging to different geographical or cultural regions.
df=pd.read_csv(r"C:\Users\chandana\OneDrive\Pictures\Desktop\datasets\CSV files\2015.csv")
# --- Code to Generate the Bar Plot ---
# 1. Calculate Average Happiness Score per Region
average_happiness_by_region = df.groupby('Region')['Happiness Score'].mean()
# 2. Create the Bar Plot
plt.figure(figsize=(10, 6))  # Adjust figure size for better readability
plt.bar(average_happiness_by_region.index, average_happiness_by_region.values)
# 3. Customize the Plot
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.title('Average Happiness Score by Region')
plt.xticks(rotation=45, ha="right")  # Rotate x-axis labels for better visibility
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
# 4. Show the Plot
plt.show()
```

```python
#HISTOGRAM
#How are the Standard Errors distributed for the Happiness Scores? 
#The variability of the standard errors associated with the reported Happiness Scores across all the countries in your dataset.

df = pd.read_csv(r"C:\Users\chandana\OneDrive\Pictures\Desktop\datasets\CSV files\2015.csv")
# --- Code to Generate the Histogram of Standard Errors ---
# Create the Bar Plot
plt.figure(figsize=(8, 6))  # Adjust figure size as needed
plt.hist(df['Standard Error'], bins=10, color='lightcoral', edgecolor='black')
# Customize the plot
plt.xlabel('Standard Error of Happiness Score')
plt.ylabel('Frequency (Number of Countries)')
plt.title('Distribution of Standard Errors of Happiness Scores')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
# Show the plot
plt.show()
```
